```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_name = r'D:\python\Data Science\Python\E-Commerce Sales Analysis\data\datasets\Sale Report.csv'
df = pd.read_csv(file_name)
```

```python
# Display the first few rows to understand the structure
print("First 5 rows of the dataset:")
print(df.head())

# Basic info about the dataset
print("\nBasic Info:")
df.info()

# Check the summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# 1. Visualize stock trends over the index 
df_sales_trend = df.groupby('index')['Stock'].sum()

# Plot the stock trends over index
plt.figure(figsize=(10, 6))
plt.plot(df_sales_trend.index, df_sales_trend.values, color='blue', marker='o', linestyle='-', markersize=5)
plt.title('Stock Trends Over Time (Index)', fontsize=16)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Total Stock', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Bar plot
df_sales_trend = df.groupby('index')['Stock'].sum()
plt.figure(figsize=(10, 6))
plt.bar(df_sales_trend.index, df_sales_trend.values, color='blue')
plt.title('Stock Trends Over Time (Index)', fontsize=16)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Total Stock', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# 2. Analyze relationships between different features (Category, Size, Color)

# Sales by Category
df_category_sales = df.groupby('Category')['Stock'].sum().reset_index()

# Plot the total sales for each category
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Stock', data=df_category_sales, palette='viridis')
plt.title('Stock by Product Category', fontsize=16)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Total Stock', fontsize=12)
plt.xticks(rotation=45)
plt.show()

# Sales by Size
df_size_sales = df.groupby('Size')['Stock'].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Size', y='Stock', data=df_size_sales, palette='Blues')
plt.title('Stock by Product Size', fontsize=16)
plt.xlabel('Product Size', fontsize=12)
plt.ylabel('Total Stock', fontsize=12)
plt.show()

# Sales by Color
df_color_sales = df.groupby('Color')['Stock'].sum().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Color', y='Stock', data=df_color_sales, palette='Set2')
plt.title('Stock by Product Color', fontsize=16)
plt.xlabel('Product Color', fontsize=12)
plt.ylabel('Total Stock', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Identify patterns or anomalies in the sales data (best-selling products)

df_product_sales = df.groupby('SKU Code')['Stock'].sum().reset_index().sort_values(by='Stock', ascending=False)

# Plot the top 10 best-selling products
plt.figure(figsize=(8, 6))
sns.barplot(x='SKU Code', y='Stock', data=df_product_sales.head(10), palette='coolwarm')
plt.title('Top 10 Best-Selling Products (SKU Code)', fontsize=16)
plt.xlabel('Product (SKU Code)', fontsize=12)
plt.ylabel('Total Stock', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
