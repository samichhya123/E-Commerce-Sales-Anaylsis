```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('D:\python\Data Science\Python\E-Commerce Sales Analysis\data\datasets\Sale Report.csv')

# Assuming each 100 rows represent a month
df['Month'] = (df['index'] // 100) + 1  # Group every 100 rows into a month number

# Sales per Month
monthly_sales = df.groupby('index')['Stock'].sum().reset_index()

# Plot Monthly Sales Trends
plt.figure(figsize=(10, 6))
sns.lineplot(x='index', y='Stock', data=monthly_sales, marker='o')
plt.title('Monthly Sales Trends', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Stock', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Create a new 'Season' column based on the month number
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['Season'] = df['Month'].apply(get_season)

# Sales per Season
seasonal_sales = df.groupby('Season')['Stock'].sum().reset_index()

# Plot Seasonal Sales Trends
plt.figure(figsize=(8, 6))
sns.barplot(x='Season', y='Stock', data=seasonal_sales, palette='Set3')
plt.title('Seasonal Sales Trends', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Total Stock', fontsize=12)
plt.show()
```

```python
# Assuming every 12 months make up a year in the dataset
df['Year'] = (df['Month'] - 1) // 12 + 1

# Sales per Year
yearly_sales = df.groupby('Year')['Stock'].sum().reset_index()

# Plot Yearly Sales Trends
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Stock', data=yearly_sales, palette='Blues_d')
plt.title('Yearly Sales Trends', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Stock', fontsize=12)
plt.show()
```

```python
# Sales by Category over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Month', y='Stock', hue='Category', data=df, marker='o')
plt.title('Sales by Product Category Over Time', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Stock', fontsize=12)
plt.grid(True)
plt.show()
```
